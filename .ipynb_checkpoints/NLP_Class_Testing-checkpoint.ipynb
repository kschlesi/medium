```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy_utils import database_exists, create_database
import psycopg2
import NLP
```

```python
# connect to postgresql db
username = 'kimberly'
dbname = 'medium'

dbe = create_engine('postgres://%s@localhost/%s'%(username,dbname))
```

```python
# get df, drop missing data
df = pd.read_sql('articles', dbe, index_col='postid')
df = df.dropna(axis=0,how='any')
print(df.shape)
df.head()
```

<h3>Testing NLProcessor...</h3>

```python
parArticleText = NLP.NLProcessor(list(df.text))
```

```python
parArticleText.get_text(ttype='raw',output_type='dataframe')
```

```python
parArticleText.process_text(break_on=['/n'], to_stem=False, init_split_on='database', origdb=list(df.origdb))
```

```python
textdf = parArticleText.get_text(ttype='tokenized',output_type='dataframe')
print(textdf.shape)
textdf.head()
```

```python
sentArticleText = NLP.NLCorporizer(list(df.text))
```

```python
sentArticleText.process_text(break_on=['.'], init_split_on='database', origdb=list(df.origdb))
sentdf = sentArticleText.get_text(ttype='tokenized',output_type='dataframe')
print(sentdf.shape)
sentdf.head()
```

```python
sentArticleText.flatten_text()
```

```python
flatdf = sentArticleText.get_text(ttype='flat',output_type='dataframe')
print(flatdf.shape)
flatdf.head()
```
