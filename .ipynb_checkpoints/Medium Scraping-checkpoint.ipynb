```python
myurls = ['https://blog.medium.com/welcome-to-the-medium-api-3418f956552',
          'https://medium.com/the-mission/in-1995-this-astronomer-predicted-the-internets-greatest-failure-68a1c3927e46',
          'https://medium.freecodecamp.org/9-popular-online-courses-that-are-gone-forever-and-how-you-can-still-find-them-a716ddf8e1d9',
          #'https://medium.com/@heyellieday/scraping-medium-posts-f94ed9f0696e'
         ]

hilites = []
for myurl in myurls:
    print(myurl)
    browser.get(myurl)
    sleep(5)
    try:
        hilites.append(browser.find_element_by_css_selector('.is-other').text)
        
    except:
        hilites.append('')
    
#browser.close()    
```
