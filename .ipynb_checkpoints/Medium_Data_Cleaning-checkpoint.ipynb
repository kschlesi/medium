```python
# connect to postgresql db
username = 'kimberly'
dbname = 'medium_test_7'

dbe = create_engine('postgres://%s@localhost/%s'%(username,dbname))
print(dbe.url)
```

```python
# get df 

df = pd.read_sql('articles', dbe, index_col='postid')
df.head()
```

```python
# check out the Title/Text column
df.highlight
```

```python
# where is data missing?
print(df.shape)
for i in df.columns:
    print(i,df[i].count())

df_all = df.dropna(axis=0,how='any')
print(df_all.shape)
```

```python
# dataframe of records missing any values
df_missing = df[df.isnull().any(1)]
print(df_missing.shape)
df_missing.head()
```

```python
print(df_missing)
```

```python
# count in each column the number that are none
total_count = df.shape[0]
missing_counts = []
for i in df_missing.columns:
    mcount = df_missing[i][df_missing[i].isnull()].shape[0]
    missing_counts.append(mcount)
    
fig, ax = plt.subplots()

xtick = range(len(df_missing.columns))
mplot = [m/total_count for m in missing_counts]
rects1 = ax.bar(xtick, 
                mplot, 
                color='r')
ax.set_ylabel('percent missing')
ax.set_title('missing records by field')
ax.set_xticks(xtick)
ax.set_xticklabels(df_missing.columns)

plt.show()
```

<h3>Creating Final Database</h3>

```python
# pull in df for each db
username = 'kimberly'
dbnames = ['medium_test_%s'%i for i in range(6,11)]

engines = [create_engine('postgres://%s@localhost/%s'%(username,db)) for db in dbnames]
dfs = [pd.read_sql('articles', e, index_col='postid') for e in engines]
dfrows = [df.shape[0] for df in dfs]
print(dfrows)
print(sum(dfrows))
```

<b>\n issue is weird.... instead save a new npar field (later) and orig_db as well (now)</b>

```python
# fix \n problem in text field
# NO DON'T
for df in dfs:
    if len(df.text[0].split('\n')) != df.npar[0]:
        print(df.shape[0])
```

```python
# add orig_db field to each df

for i,df in enumerate(dfs):
    df['origdb'] = i
    
```

```python
# concat all dbs (but remove duplicates at the end of db3 - beginning of db4)
#[i in dfs[3].index for i in dfs[4].index]
ixs = [ix for ix,b in enumerate([i in dfs[4].index for i in dfs[3].index]) if b==True]
print(ixs[0:3])
# remove from 2629 on

dfs[3] = dfs[3][0:ixs[2]]

df_all = pd.concat(dfs)
df_all.head()
```

```python
# create separate dfs of url/date in order of dfU

# load in current urls
fileU = 'medium_urls_unique.txt'
fileD = 'medium_urls/medium_urls_dates.txt'
# here are the files
dfU = pd.read_table(fileU,header=None)
dfD = pd.read_table(fileD,header=None,delimiter=',')

# load in scrapy urls + dates
print(dfU.shape)
print(dfD.shape)

# join on url
dfUD = pd.merge(dfU, dfD, how='left', left_on=dfU.columns[0], right_on=dfD.columns[0])
dfU.head()  # still in order
dfUD.head() # still in order but with multiples........
```

```python
# find UNIQUE url/dates (keep oldest popular date) and create a new column with urlID

dfUD_unique = dfUD.groupby(dfUD.columns[0],as_index=False,sort=False).min()
dfUD_unique.columns = [['url','popdate']]
dfUD_unique['urlid'] = dfUD_unique['url'].map(lambda x : x.split('-')[-1])
dfUD_unique.set_index('urlid',inplace=True)
dfUD_unique.head()
```

```python
# merge the df of scraped data with the matching UNIQUE dfD urls/dates (merge on postid)

print(dfUD_unique.shape)
print(df_all.shape)

dfOld = pd.merge(df_all,dfUD_unique,how='right',left_index=True,right_index=True,sort=False)
print(dfOld.shape)
dfOld.head()
```

```python
# format dataframe for db

dfOld = dfOld[['title','popdate','url','userid','username','highlight','nlikes','ncomments','ntags','origdb','tags','text','npar']]
dfOld.index.name = 'postid'
print(dfOld.shape)
dfOld.head()
```

```python
# create new db and put df in it
# pull in df for each db
username = 'kimberly'
newdb = 'medium'

engine = create_engine('postgres://%s@localhost/%s'%(username,newdb))
if not database_exists(engine.url):
    create_database(engine.url)

dfOld.to_sql('articles',engine,if_exists='replace')
```
